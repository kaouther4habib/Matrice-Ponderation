```
import numpy as np
import matplotlib.pyplot as plt
```

```
def h(x):
  w1 = 0
  w2 = 1
  y = w1 + w2 *x
  return y
```

```
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 2, 3, 4, 5])
plt.scatter(x,y)
x1 = np.array([0, 5])

plt.plot(x, h(x))
```

```
def compute_loss(x, y, h):
  mse = np.sum((h(x) - y)**2) / (2 * x.shape[0])
  return mse
```

```
compute_loss(x, y, h)
```

```
w2_values = [i for i in np.arange(-10, 10, 0.1)]
y_loss = []
for w2_value in w2_values:
  def f(x):
    w1 = 0
    w2 = w2_value
    y = w1 + w2 * x
    return y
  y_loss.append(compute_loss(x, y, f))

plt.plot(w2_values, y_loss)
```

```
w1_values = np.linspace(-10, 10, 50)
w2_values = np.linspace(-10, 10, 50)
loss = np.zeros((w1_values.shape[0], w2_values.shape[0]))

for i, w1_value in enumerate(w1_values):
  for j, w2_value in enumerate(w2_values):
    def f(x):
      w1 = w1_value
      w2 = w2_value
      y = w1 + w2 * x
      return y
    loss[i,j] = compute_loss(x, y, f)
plt.contour(w1_values, w2_values, loss, 2000, cmap=plt.cm.jet)
cbar = plt.colorbar()
```

```
import math

def sigmoid(x):
  return 1 / (1 + math.exp(-x))

def h(x1, x2):
  w1 = -2
  w2 = 2
  y = x1*w1 + x2*w2
  return sigmoid(y)

def line_fn(x2):
  w1 = -2
  w2 = 2
  x1 = -(w2/w1) * x2
  return x1
```

```
x1_values = np.linspace(-10, 10, 50)
x2_values = np.linspace(-10, 10, 50)
y_values = np.zeros((x1_values.shape[0], x2_values.shape[0]))

for i, x1_value in enumerate(x1_values):
  for j, x2_value in enumerate(x2_values):
    y_values[i,j] = h(x1_value, x2_value)

plt.imshow(y_values, extent=(x1_values.min(), x1_values.max(), x2_values.max(), x2_values.min()))
cbar = plt.colorbar()
x = np.array([-5, 5])
plt.plot(x, line_fn(x))
```
